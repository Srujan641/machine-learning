```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df= pd.read_csv("price.csv")
df
```

```python
x=np.array(df.price)
y=np.array(df.area)
x
```

```python
y
```

```python
plt.xlabel("price")
plt.ylabel("area")
plt.scatter(x,y)
```

```python
reg=linear_model.LinearRegression() # objest in the call linear regresiiion is created
reg.fit(df[['area']],df.price) # .fit is used to train the model with the data set
#always the first dta set passed must be a 2d array
```

```python
reg.predict([[1000]])
#predics the price of the land based on the givem area
```

```python
plt.scatter(x,y)
plt.plot(reg.predict(df[['area']]),df.area) #command to plot a graph after predicting the value 
```
